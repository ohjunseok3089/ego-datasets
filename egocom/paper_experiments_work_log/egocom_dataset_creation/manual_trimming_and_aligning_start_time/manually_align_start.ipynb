## Simple script to noisily align the videos by hand (just get them to a starting place within a few seconds.)

```python
from __future__ import print_function, absolute_import, division, unicode_literals, with_statement # Python 2 compatibility

import os
import subprocess
import numpy as np 
```

```python
data_loc = "/media/cgn/ seagate1tb/egocentric_conversations/"
```

# Generate code to be run for day1-day2 video

```python
for dayfolder in sorted([v for v in os.listdir(data_loc) if os.path.isdir(data_loc+v)]):
    base = data_loc + dayfolder + "/"
    for cfolder in sorted([z for z in os.listdir(base) if os.path.isdir(base+z)]):
        path = base + cfolder + "/"
        file_matrix = []
        for devicefolder in sorted([w for w in os.listdir(path) if os.path.isdir(path+w)]):
            subpath = path+devicefolder + "/"
            if os.path.isdir(subpath) and len(os.listdir(subpath)) > 0:
                files = [subpath+y for y in os.listdir(subpath) if y[0] != '.']
                files.sort(key=lambda x: os.stat(os.path.join(subpath, x)).st_mtime)
                file_matrix.append(files)

        if len(file_matrix) > 0:
            print("#", path)
            print()

            files_per_segment = [[row[i] for row in file_matrix] for i in range(len(file_matrix[0]))]

            for j, file_list in enumerate(files_per_segment):
                if j == 0:
                    print("s = [10.0, 11.1, 12.2] # How much to remove from front of clip")
                elif j == 1:
                    print("s = [ss - min(s) for ss in s] # How much to remove from front of clip")
                for i, f in enumerate(file_list):
                    cmd = "ffmpeg -ss 00:00:01 -i {f}.MP4 -c copy {f}_frontcut.MP4".format(f=f)
                    print('''cmd = "ffmpeg -ss 00:00:{{s}} -i '{f}.MP4' -map_channel 0.1.1 -map_channel 0.1.0 -c:v copy '{f}_frontcut.MP4' -y".format(s=s[{idx}])'''.format(f=f[:-4],idx=i))
        #             print('cmd = "vlc {f}"'.format(f=f))
                    print('out = subprocess.getoutput(cmd)')

                print()

            print("*"*80, "\n")
```

# Rename all files

```python
data_loc = "/media/cgn/ seagate1tb/egocentric_conversations/"
d = 0
for dayfolder in sorted([v for v in os.listdir(data_loc) if (os.path.isdir(data_loc+v) and v[:3] == 'day')]):
    d += 1
    base = data_loc + dayfolder + "/"
    c = 0
    for cfolder in sorted([z for z in os.listdir(base) if (os.path.isdir(base+z) and z[:12] == 'conversation')]):
        c += 1
        path = base + cfolder + "/"  
        file_matrix = []
        for devicefolder in sorted([w for w in os.listdir(path) if (os.path.isdir(path+w) and w[:6] == 'device')]):
            subpath = path+devicefolder + "/"
            if os.path.isdir(subpath) and len(os.listdir(subpath)) > 0:
                files = [subpath+y for y in os.listdir(subpath) if y[0] != '.']
                files.sort(key=lambda x: os.stat(os.path.join(subpath, x)).st_mtime)
                file_matrix.append(files)

        if len(file_matrix) > 0:
            print()

            files_per_segment = [[row[i] for row in file_matrix] for i in range(len(file_matrix[0]))]
            
            part = 0
            for file_list in files_per_segment:
                part += 1
                p=0
                for f in file_list:
                    p += 1
                    cmd = "mv '{f}' '{l}day_{d}__con_{c}__person_{p}_part{part}.MP4'".format(f=f,d=d,c=c,p=p,part=part,l=data_loc)
                    out = subprocess.getoutput(cmd)
```
